## 물리 배경

정상파의 경우 다음 공식이 적용됩니다.
$v=\lambda f$

여기서 $v$는 파동의 속도(m/s), $\lambda$는 파장(m), $f$는 주파수(Hz)를 나타냅니다. 팽팽한 줄에서의 파동 속도는 다음 공식을 사용하여 계산할 수도 있습니다.
$v=\sqrt{\frac{F_s}{\rho_L}}$

이 공식에서 $v$는 다시 파동의 속도(m/s), $F_s$는 줄의 장력(N), $\rho_L$은 줄의 선밀도(kg/m)를 나타냅니다. 선밀도는 줄의 질량을 길이로 나누어 결정할 수 있습니다. 두 공식을 같게 놓으면 다음과 같습니다.
$\sqrt{\frac{F_s}{\rho_L}}=\lambda f$

실험에서의 변수는 파장과 줄의 장력입니다. 이 실험 내에서 이 두 양을 서로 연관시킴으로써 위의 관계를 다음과 같이 표현할 수 있습니다.
$\lambda^2=\frac{F_s}{\rho_L f^2}$

다이어그램에서 줄의 장력에 대해 파장의 제곱을 도표화하면 비례 관계가 나타납니다. 그래프의 기울기는 $\frac{1}{\rho_L f^2}$과 같으며, 이로부터 주파수를 계산할 수 있습니다. 그래프의 예는 {numref}`그림 {number} <demo77_fig2>`에 나와 있습니다.

## 후속 활동
다른 고무줄(선밀도가 다른 것)이나 다른 모델 또는 브랜드의 전동 칫솔을 사용하면 추가적인 연구를 자극할 수 있습니다.


````python
# 필요한 라이브러리 임포트
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# 측정값 입력
mass = np.array([..,..,..,..,..]) # 물체의 질량
wavelength = np.array([..,..,..,..,..]) # 결정된 파장

# 측정값 플로팅
plt.figure
plt.plot(mass, wavelength, 'o')
plt.xlabel('질량')
plt.ylabel('파장')

# 축 범위 설정
plt.xlim(0,1.1*max(mass))
plt.ylim(0,1.1*max(wavelength))
plt.grid()
plt.show()
````

````python
# 데이터 피팅
def function(mass,a):
    return a*np.sqrt(mass*9.81)

var, cov = curve_fit(function, mass, wavelength)

print('a의 값은:', var[0])

# 데이터와 피팅 결과 플로팅
m_test = np.linspace(0.9*min(mass), 1.1*max(mass), 100)
w_test = function(m_test, var[0])

plt.figure
plt.plot(mass, wavelength, 'o')
plt.plot(m_test, w_test,'r--')
plt.xlabel('질량')
plt.ylabel('파장')

plt.xlim(0,1.1*max(mass))
plt.ylim(0,1.1*max(wavelength))
plt.grid()
plt.show()
````
