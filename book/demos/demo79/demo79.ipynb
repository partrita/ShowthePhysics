## 물리 배경
한쪽이 막힌 관에서 기본 주파수는 다음과 같습니다.
$f=\frac{v}{\lambda}=\frac{v}{4 l^{\prime}}$

여기서 $l^{\prime}$이라고 적은 이유는 마디가 튜브의 입구에 정확히 위치하는 것이 아니라 약간 바깥쪽에 위치하기 때문입니다. 즉:
$l^{\prime}=l+\Delta l$

여기서 $\Delta l$은 계통 오차(개구단 보정)입니다. 이 편차 때문에 여러 길이에 대한 측정값이 필요한 것입니다.

## 코드 셀

````python
#라이브러리 로딩
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

#측정값 
L = np.array([9.9, 19.8, 25.0, 29.9, 37.3])*1e-2 #cm 단위 입력
u_L = np.ones(len(L))*0.001 #m
f_0 = np.array([860, 439, 352, 293, 233]) #Hz
u_f_0 = np.ones(len(f_0))*3 #Hz 5, 2, 1, 3

lambda_f_0 = 4*L
u_lambda_f_0 = 4*u_L
print("데이터를 성공적으로 로드했습니다.")
````

````python
#데이터 플로팅
plt.figure()
plt.errorbar(L,f_0,xerr=u_L,yerr=u_f_0,linestyle='none')
plt.plot(L,f_0,'k.')
plt.xlabel('$L$ (m)')
plt.ylabel('$f$ (Hz)')
plt.show()
````

````python
#함수 피팅
def funcfit(x,v,dL):
    return v/(x+dL)

var, cov = curve_fit(funcfit,lambda_f_0,f_0,p0=[343,2e-2])
print(var)
print()
print('공기 중 소리의 속도는 다음과 같습니다:', round(var[0],0),'+/-',round(np.sqrt(cov[0,0]),1),'m/s')
````

````python
#데이터와 피팅 함수 플로팅
x = np.linspace(0.9*min(lambda_f_0),1.1*max(lambda_f_0))
y = funcfit(x,var[0],var[1])

plt.figure()
plt.plot(x,y,'r-')
plt.plot(lambda_f_0,f_0,'k.')
plt.xlabel('$\lambda$ (m)')
plt.ylabel('$f$ (Hz)')
plt.xlim([0.95*min(lambda_f_0),1.05*max(lambda_f_0)])
plt.show()
````
