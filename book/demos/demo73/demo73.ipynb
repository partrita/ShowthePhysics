## 물리 배경
모든 측정은 일정한 정밀도로만 수행될 수 있습니다. 때로는 정밀도가 장비에 의해 결정되고, 때로는 측정하려는 속성에 의해 결정됩니다. 과학에서 답은 항상 연관된 불확실성과 함께 주어지며, 반복된 측정이 연관된 불확실성을 가진 값 내에 떨어질 확률은 68%입니다 ($P(\bar{x}-\mu_x<x<\bar{x}+\mu_x)=0.68$). 중력 가속도 결정에서와 같이 여러 양이 역할을 할 때 각 측정이 기여합니다. 이 방정식의 최종 불확실성 값은 다음과 같이 주어집니다.

$$
    (\frac{\mu_g}{g})^2 =  (\frac{\mu_H}{H})^2 + 4 (\frac{\mu_{\Delta t}}{\Delta t})^2
$$

불확실성은 항상 유효 숫자 한 자리로만 제공됩니다. 최종 답은 동일한 소수점 자릿수로 제공됩니다 (예: 9.8 ± 0.3 m/s²).
