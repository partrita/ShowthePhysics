## 소개
이 놀라운 시연은 전구의 PTC(정온도계수, Positive Temperature Coefficient) 동작을 시각적으로 보여줍니다. 관찰 결과를 설명하려면 많은 [상호 피드백 사고](../../Pedagogy/BackAndForthThinking.md)가 필요합니다. 이 시연은 다양한 수준의 학생들에게 영감을 줄 수 있으며, 특히 직렬 회로에서의 저항, 전류 및 전압에 대한 학생들의 이해도를 확인하는 데 적합합니다.

## 장비
- 전원 공급 장치
- 연결 재료를 포함한 4~5개의 동일한 전구
- 선택 사항: FLIR 카메라

## 준비
전구들이 직렬로 연결되어 있음을 명확히 보여주도록 연결 상태를 설정합니다. 칠판에 회로도를 그립니다.

```{figure} demo85_figure1.JPG
---
width: 100%
align: center
---
시연 중에 음극 단자의 연결을 왼쪽에서 오른쪽으로 계속 이동시켜, 처음에는 전구 하나만 켜지고 마지막에는 모든 전구가 연결되도록 합니다. 오른쪽에는 대안적인 방법이 나와 있습니다.
```

```{tip}
영상에서처럼 회로의 일부를 단락(short circuit)시킬 수 있습니다. 단락된 전구는 불이 들어오지 않습니다.
```

## 절차
전압을 일정하게 유지하면서 전원 공급 장치에 전구를 하나씩 더 연결하며 시연을 시작합니다. 학생들은 관찰 내용을 기록합니다. 그런 다음 관찰 내용을 짧게 요약합니다. 가능한 관찰 내용은 다음과 같습니다:
- 연결된 단자 사이에 있는 전구들만 불이 들어온다.
- 연결된 전구의 수가 많아질수록 전구의 밝기는 어두워진다.
- 마지막으로 연결한 전구는 점차 밝아지는 반면, 이미 연결되어 있던 다른 전구들은 즉시 최종 밝기로 빛난다.
- 전구를 더 많이 연결할수록, 마지막으로 연결된 전구가 최종 밝기에 도달하는 데 걸리는 시간이 점점 더 길어진다.

```{figure} demo85_figure2.JPG
---
width: 90%
align: center

연결을 옮긴 직후의 상황. 세 번째 전구는 거의 보이지 않으며 다른 두 전구만큼 밝아지는 데 시간이 필요합니다.
```

그런 다음 관찰된 현상에 대한 가능한 설명에 대해 토론을 시작합니다. 세 번째와 네 번째 관찰 내용이 가장 흥미롭습니다. 처음 두 관찰 내용에 대한 설명을 사용하여 학생들의 사전 지식을 확인할 수 있습니다. 이 모든 관찰을 설명하려면 상호 피드백 사고가 필요하며, 심층적인 질문을 던지고 학생들이 과학적 용어를 사용하도록 유도하는 것이 중요합니다.

```{tip}
세 번째 관찰 내용을 설명하려면 전구에 불이 들어올 때 저항에 어떤 일이 일어나는지 물어보십시오. 네 번째 관찰이 아직 이루어지지 않았다면 예측해 볼 기회를 제공합니다. 학생들은 마지막으로 연결된 전구가 천천히 밝아지는 동안 전류에 어떤 일이 일어날지도 예측해 볼 수 있습니다. 전류는 감소할 것이며, 전류계로 확인할 수 있습니다.
```

## 물리 배경
설명은 세 번째 전구가 막 연결된 {numref}`그림 {number} <demo85_2>` 상황의 세 번째와 네 번째 관찰 내용과 가장 잘 연결됩니다: 전구 1과 2의 온도는 전구 3보다 높습니다. 전구 1과 2는 이미 켜져 있었고 전구 3은 그렇지 않았기 때문입니다. 따라서 전구 1과 2의 저항이 전구 3의 저항보다 높고, 결과적으로 전구 1과 2가 전압의 더 큰 몫을 가져갑니다.

연결 후 전구 3의 온도(따라서 저항)가 증가하므로, 전구 3에 걸리는 전압이 점점 더 커집니다 (그리고 전구 1과 2에 걸리는 전압은 작아지지만 밝기 감소는 육안으로 식별하기 어렵습니다). 네 번째 관찰: 더 많은 전구가 직렬로 연결될수록 전체 저항이 커지고 전류는 작아지므로, 마지막 전구의 온도가 더 천천히 올라갑니다. 이는 전구에 걸리는 전압의 증가를 늦추고 결국 밝기의 증가를 늦춥니다. 마지막으로 전류에 일어나는 일: 마지막으로 연결된 전구의 저항이 증가하므로 회로의 전체 저항이 증가하고 전류는 감소합니다.

다음 영상은 온도가 어떻게 변하는지 보여줍니다. 영상을 2배속으로 재생하는 것이 좋을 수 있습니다.

## 후속 활동
이 시연은 추가 연구, 특히 센서(또는 오실로스코프)가 있을 때 유용합니다. 아래 사진에서는 옴 저항과 전구가 직렬로 연결된 회로에 AC 신호를 인가했을 때 [이력 현상(hysteresis)](https://en.wikipedia.org/wiki/Hysteresis)을 관찰할 수 있습니다. {cite:t}`mooldijk2005`의 설명을 바탕으로 아래에 시뮬레이션을 만들었습니다. 오른쪽 상단의 로켓({fa}`rocket`)을 클릭하십시오.

``` {figure} demo85_figure3.png
---
width: 50%
---
오실로스코프로 보여주는 이력 현상. 신호 1은 5Hz로 설정된 함수 발생기에서, 신호 2는 (전구와 직렬로 연결된) 저항기 양단에서 측정한 것입니다. 오실로스코프는 (X-Y) 모드로 설정되었습니다.
```

````python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# 아드 몰데이크(Ad Mooldijk) 제작 모델

# 상수
U0 = 8          # V
Tg0 = 293      # K
l = 0.0143      # m
d = 0.132e-4    # m
pi = np.pi
A = pi * (d / 2)**2  # m^2
S = l * 2 * pi * (d / 2)  # m^2
Seff = S          # m^2
rho0 = 5.5e-8     # Ohm m
c = 0.13e3        # J/kg/K
rhod = 19.3e3     # kg/m^3
m = rhod * l * A  # kg
ssigma = 5.67e-8  # W/m^2/K^4
alpha = 4.9e-3    # K^-1
dt = float(1e-5)        # s

# 시뮬레이션 실행 함수
def run_simulation(f):
    R = rho0 * l / A  # 초기 저항
    Tg = Tg0          # 초기 온도
    t = 0             # 초기 시간
    times = []
    temperatures = []
    U_x = []
    I_x = []

    # 시뮬레이션 시간 파라미터
    simulation_time = 3/f  # 총 시뮬레이션 시간 (초)
    steps = int(simulation_time / dt)

    for _ in range(steps):
        U = U0 * np.sqrt(2) * np.sin(2 * pi * f * t)
        I = U / R
        Pin = I**2 * R
        Qin = Pin * dt
        dTgin = Qin / (c * m)
        Tg += dTgin
        dPuit = ssigma * Seff * Tg**4
        dQuit = dPuit * dt
        dTguit = -dQuit / (c * m)
        Tg += dTguit
        rho = rho0 * (1 + alpha * (Tg - Tg0))
        R = rho * l / A
        t += dt
        
        # 플로팅을 위해 시간, 온도, 전압, 전류 저장
        times.append(t)
        temperatures.append(Tg)
        U_x.append(U)
        I_x.append(I)

    # 서브플롯으로 결과 플로팅
    fig, ax = plt.subplots(2, 2, figsize=(12, 6))
    
    # 시간에 따른 온도
    ax[0,0].plot(times[int(steps/3):], temperatures[int(steps/3):])
    ax[0,0].set_xlabel('시간 (s)')
    ax[0,0].set_ylabel('온도 (K)')
    ax[0,0].grid(True)
    
    # 전압 vs. 전류
    ax[0,1].plot(I_x[int(steps/3):], U_x[int(steps/3):])
    ax[0,1].set_xlabel('전류 (A)')
    ax[0,1].set_ylabel('전압 (V)')
    ax[0,1].grid(True)
    
    ax[1,0].plot(times[int(steps/3):], U_x[int(steps/3):])
    ax[1,0].set_xlabel('시간 (s)')
    ax[1,0].set_ylabel('전압 (V)')
    ax[1,0].grid(True)
    
    plt.tight_layout()
    plt.show()

# 주파수 조절 슬라이더
interact(run_simulation, f=FloatSlider(value=10, min=1, max=50, step=1, description='주파수 f (Hz)'));
````
