## 물리 배경
막대의 한쪽 끝을 치면 (예를 들어 테이블에) 막대에 다양한 파장의 진동이 생성됩니다. 막대 중간을 잡으면 그곳에 마디가 생성되어 중간에 마디가 있는 파동만 허용됩니다. 또한 주파수가 높은 진동은 더 빨리 감쇠합니다. 따라서 기본 주파수가 가장 잘 들립니다 (때로는 배음이 사라질 때까지 기다려야 합니다).

막대의 위쪽 4분의 1 지점을 잡으면 전체 파장이 포함된 진동, 즉 한 옥타브 높은 음이 기본음으로 맞게 됩니다.

사포로 문지르면 막대는 횡방향과 종방향으로 모두 진동합니다. 이는 막대가 '노래'하게 만든 다음 감쇠시킴으로써 쉽게 시연할 수 있습니다. 종방향 감쇠는 막대의 양 끝을 잡음으로써 일어나고, 횡방향 감쇠는 막대를 움켜쥠(너비를 둘러싸는 방식)으로써 일어납니다. 막대는 양쪽 방향으로 진동합니다.
기본음의 경우, 중간에 마디가 있고 끝부분에 배마디가 있습니다. 따라서 파장은 막대 길이의 두 배입니다. 해당 주파수는 다음과 같이 계산할 수 있습니다.

$$
f = \frac{n \cdot v_{Al}}{2l}
$$ (eq:demo81_1)

막대에서의 소리의 속도는 다음과 같이 계산할 수 있습니다.

$$ v = \sqrt{\frac{E}{\rho}} $$ (eq:demo81_2)

여기서 $E$는 사용된 재료의 탄성 계수(알루미늄의 경우 71 GPa)이고 $ρ$는 막대의 밀도입니다. 반대로, 측정된 주파수를 사용하여 재료의 탄성 계수를 결정할 수 있습니다.

## Python
프릭 폴스(Freek Pols)는 이 실험을 1학년 물리학 실험 과정에서 사용하고 있습니다. 막대를 부드럽게 문지르는 대신, 그는 망치를 사용하여 막대의 한쪽 끝을 두드립니다.

그는 영률을 결정하기 위해 5개의 황동 및 알루미늄 막대를 사용합니다. Python 코드는 알루미늄 막대에 대한 그의 측정값을 포함하여 아래에 나와 있습니다. 이 값들을 변경할 수 있습니다.

````python
# 다양한 라이브러리 로딩
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# 측정값
L = np.array([39.9, 60.0, 79.9, 99.8, 119.9])*1e-2 #m
f = np.array([6408, 4274,3209, 2562, 2137]) #Hz
rho_Al = 2.7e3 #kg/m^3


# 피팅 함수 
def fitfunctie(L,a):
    return a/L

# 피팅 수행
var, cov = curve_fit(fitfunctie,f,L)

# 다양한 L에 대해 f를 계산하는 테스트 데이터셋
xtest = np.linspace(0.8*min(L),1.2*max(L),1000)
ytest = fitfunctie(xtest,var)

# 데이터 및 피팅 결과 플로팅
plt.figure()
plt.plot(L**2,1/f**2,'k.')
plt.plot(xtest**2,1/ytest**2,'r--')

# 축 설정
plt.xlim(0.9*min(L**2),1.1*max(L**2))
plt.xlabel('$L^2$ (m$^2$)')
plt.ylabel('$f^{-2}$ (Hz$^{-2}$)')

plt.show()

print('영률(Young\'s modulus)은', round(4*rho_Al*var[0]**2*1e-9), 'GPa 입니다.')
````
