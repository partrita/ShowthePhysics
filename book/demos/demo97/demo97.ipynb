## 소개
방사성 붕괴는 통계적인 과정입니다. 방사성 붕괴 과정은 학생들이 상상하기 어렵기 때문에, 이 시연은 학생들이 이를 시각화하는 데 도움을 줍니다.

한 변의 길이가 약 1.5cm인 정육면체(큐브)를 대량으로 준비합니다. 예를 들어 300개를 준비한다고 가정해 봅시다. 각 큐브의 한 면은 파란색, 두 면은 빨간색으로 칠해져 있습니다. 나머지 세 면은 아무것도 칠해져 있지 않습니다. 색칠된 큐브를 던짐으로써 방사성 붕괴의 통계적 측면을 시뮬레이션합니다. 큐브를 주사위처럼 테이블 위에 던지면 약 1/6은 파란색 면이 위로, 2/6는 빨간색 면이 위로, 3/6은 흰색(칠해지지 않은 면)이 위로 오게 됩니다.


빨간색 면이 위로 나온 큐브들을 제거하고 나머지 큐브들을 다시 테이블 위에 던집니다. 다시 빨간색 면이 위로 나온 큐브들을 골라내고 나머지 큐브들을 다시 던지는 과정을 큐브가 거의 남지 않을 때까지 반복합니다.

파란색 면이 위로 나온 블록들을 제거할 수도 있습니다. 이 경우 모든 큐브가 제거될 때까지 더 많은 횟수의 던지기가 필요할 것입니다.

## 장비
좋은 통계적 근사를 위해 처음에 300개 정도의 큐브가 있으면 충분합니다.

## 준비
단면이 정사각형인 긴 막대기를 잘라 300개의 큐브를 만듭니다. 한 면은 파란색, 두 면은 빨간색으로 칠합니다. 톱질하여 자르고 나면 큐브의 세 면은 아무것도 칠해지지 않은 상태가 됩니다.

## 절차
* 처음 큐브의 개수를 셉니다. 이 숫자를 $N(0)$으로 표시합니다. 문자 $N$은 남은 큐브의 수(Number)를 의미합니다. 숫자 0은 0번 던졌음을 의미합니다.
* 모든 큐브를 처음으로 던집니다. 빨간색 면이 위로 나온 모든 큐브를 제거합니다. 남은 큐브의 수를 $N(1)$이라고 합니다. 이는 한 번 던진 후 남은 큐브의 수입니다.
* 남은 큐브들을 두 번째로 던집니다. 다시 빨간색 큐브들을 제거합니다. 이제 $N(2)$개의 큐브가 남습니다.
* 큐브가 다 떨어지기 전까지 약 10번 정도 이 과정을 반복합니다.
* 던진 횟수 $n$에 대한 남은 큐브의 수 $N(n)$ 그래프를 그립니다.
* $N(n)$과 초기 개수 $N(0)$ 사이의 관계를 보여주는 공식을 고안해 봅니다.
* 이 공식과 관련된 그래프를 그립니다.
* 실제로 샌 횟수 그래프와 공식에 의한 그래프 사이의 일치 정도에 대해 의견을 나눕니다.

* 빨간색 면 대신 파란색 면이 나온 큐브를 제거한다면 그래프가 어떻게 될지 예측해 봅니다.
* 실제로 횟수를 세어 보고 그래프가 이론적 그래프와 얼마나 잘 일치하는지 확인합니다.

시간에 따른 방사성 핵종 수의 그래프를 보여줍니다. (또는 방사성 물질의 방사능 활동도 그래프.) 그 그래프에 해당하는 공식은 다음과 같습니다.
$N(t)=N(0) (1/2)^{(t/τ)}$
즉:
*그래프의 모양은 남은 큐브 수의 횟수 그래프와 매우 유사합니다.*

## 물리 배경
* *큐브*의 경우, 남은 큐브의 수는 처음 시작한 개수와 빨간색 면이 나올 확률에 따라 달라집니다.
* *원자핵*의 경우, 남은 원자핵의 수는 처음 시작한 핵의 수와 핵이 붕괴할 확률에 따라 달라집니다.

## 후속 활동
방사성 과정에서 방사성 물질 **A**는 물질 **B**로 붕괴합니다. 이것은 큐브(물질 **A**)를 던지는 것으로 시뮬레이션되었습니다. 빨간색 큐브(물질 **B**)는 제거되었습니다.
물질 **B** 자체가 방사성을 띠어 물질 **C**로 붕괴할 가능성도 충분합니다. 즉, **A** → **B** → **C** 과정입니다. 따라서 한편으로는 **A**의 붕괴로 인해 핵 **B**의 수가 증가하고, 다른 한편으로는 **C**로의 붕괴로 인해 핵 **B**의 수가 감소합니다. 만약 물질 **C**가 안정적이라면, **C**의 양은 계속해서 증가하기만 할 것입니다.

두 종류의 던지기를 병행하여 이 상황을 시뮬레이션할 수도 있습니다. 첫 번째 종류는 위에서 설명한 시리즈입니다. 두 번째는 다음과 같이 진행됩니다.
* 제거된 빨간색 큐브(물질 **B**)를 항상 모아두되, 이들도 함께 던집니다.
* 이들이 파란색 면(물질 **C**)이 나오면 골라냅니다.
* 골라낸 빨간색 큐브의 수는 물질 **A**의 붕괴로 인해 늘어나지만, 던졌을 때 파란색 면이 나와서(물질 **C**로 변함) 줄어들기도 합니다.
* 파란색 큐브의 수는 계속해서 증가하기만 할 것입니다.

```{tip}
이 시연은 주사위로도 할 수 있습니다. 6면보다 적은 면을 가진 주사위들도 있습니다.

엑셀 등을 사용하여 이 '붕괴 과정'의 컴퓨터 시뮬레이션을 만드는 것은 꽤 쉽습니다. 아래에 Python 시뮬레이션을 제공합니다. 화면 상단의 로켓 아이콘을 클릭하여 직접 실행하거나 코드를 변경해 볼 수 있습니다.
```



````python
# 필요한 라이브러리 로딩
import numpy as np
import matplotlib.pyplot as plt

# 초기 조건 설정
t_run = 1000 #s
dt = 1 #s

# 원자핵 A의 반감기
tau_A = 10 #s

# 원자핵 B의 반감기
tau_B = 30 #s

# 원자핵 A의 초기 개수
N_A_0 = 15000


# 이 코드는 수정할 필요가 없습니다.
N_total = int(t_run/dt)
N_A = np.zeros(N_total)
N_A[0] = N_A_0
N_B = np.zeros(N_total)
N_C = np.zeros(N_total)

t = np.zeros(N_total)

L_A = np.log(2)/tau_A
L_B = np.log(2)/tau_B


# 남은 핵의 수 계산
for i in range(len(N_A)-1):
    A_A = N_A[i]*L_A*np.exp(-L_A*dt)
    N_A[i+1] = N_A[i] - A_A*dt

    A_B = N_B[i]*L_B*np.exp(-L_B*dt)
    N_B[i+1] = N_B[i] - A_B*dt + A_A*dt

    N_C[i+1] = N_C[i] + A_B*dt

    t[i+1] = t[i] + dt

# 데이터 플로팅
plt.figure()
plt.plot(t,N_A,'k.',label='핵 A의 수')
plt.plot(t,N_B,'r.',label='핵 B의 수')
plt.plot(t,N_C,'b.',label='핵 C의 수')
plt.xlabel('시간')
plt.ylabel('핵의 수')
plt.legend()
plt.show()
````
