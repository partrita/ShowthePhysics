## 데이터 저장 및 처리
배열에 데이터를 '저장'할 수 있습니다. 해당 배열의 각 값은 처리, 플로팅, 인쇄 등을 할 수 있습니다. 아래 예에서는 $a$라는 배열이 있습니다. 일부 값이 저장됩니다. 이 값을 인쇄하고 그중 하나를 변경한 다음 값을 다시 인쇄합니다(오른쪽 상단 모서리에 있는 {fa}`rocket`을 클릭하고 라이브 코드를 클릭하여 코드 셀을 실행할 수 있습니다).

```python
import numpy as np
a = np.array([0.0, 1.0, 1.2, 0.8, 0.5]) #일부 값으로 배열 만들기.
print(a)                                #값 인쇄.
a[1] = 2.0                              #두 번째 요소의 값 변경.
print(a)                                #값 인쇄.
```

위에서 보듯이, 값을 인쇄하려면 `print` 명령을 사용합니다.

데이터로 채워진 배열을 사용하여 일부 통계 계산을 수행할 수 있습니다:

```python
a_mean = np.mean(a)                          #a의 평균값 계산
a_std = np.std(a,ddof=1)                     #a의 표준 편차 계산
a_unc = np.std(a,ddof=1)/np.sqrt(len(a))     #a의 표준 오차 계산

print("a의 평균값은: ",a_mean)         #a의 평균값 인쇄
print("a의 표준 편차는: ",a_std)  #a의 표준 편차 인쇄
print("a의 표준 오차는: ",a_unc)      #a의 표준 오차 인쇄

#더 깔끔하게:
print("a의 표준 오차는: ", round(a_unc,1))      #a의 표준 오차 인쇄
```

## 플로팅

종종 데이터의 시각적 표현을 원합니다. 파이썬에서는 matplotlib 라이브러리를 사용하여 데이터를 쉽게 플로팅할 수 있습니다. 이 라이브러리는 다양한 플롯을 지원하며, 아래에서 몇 가지를 보여드리겠습니다. 그러나 더 많은 예제는 <a href="https://matplotlib.org/stable/tutorials/introductory/pyplot.html" target="_blank">여기</a>에서 찾을 수 있습니다.

### 히스토그램

히스토그램은 데이터 세트에서 값이나 특정 범위가 얼마나 자주 발생하는지 보여줍니다.

```python
import matplotlib.pyplot as plt #matplotlib 라이브러리 가져오기

# 정규 분포를 가진 데이터 세트 만들기
data = np.random.normal(0,5,1000)
plt.figure(num='histfig')   # 새 그림 만들기

plt.hist(data,bins='auto',density=1, label='발생') # 데이터 플로팅

plt.xlabel('값')        #x축 레이블
plt.ylabel('확률')   #y축 레이블
plt.legend()                #범례 포함
plt.show()                  #플롯 표시
```

### 산점도

데이터 세트의 측정값은 산점도에 표시될 수 있습니다. 점만 플로팅되고(여기서는 검은색 `k.`) 점 사이에 선이 그려지지 않습니다. 아래는 산점도를 만드는 방법의 예입니다.


```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [1.1, 4.3, 9.2, 16.1, 25.1]

plt.figure(num='figscatterplot')

plt.plot(x,y,'k.', label='측정값 ')

#시각 자료 향상
plt.xlabel('입력 전압 [V]')
plt.ylabel('출력 전압 [V]')
plt.legend()
plt.grid()
plt.xlim(0,6)

plt.show()

```

## 곡선 맞춤

최적선을 찾기 위해 **scipy** 라이브러리의 `curve_fit`을 사용할 수 있습니다(최소 제곱 회귀를 사용합니다). 이 함수는 세 가지 입력 값이 필요합니다: 맞춰야 할 함수, $x$-값 및 $y$-값. $x$ 및 $y$ 값은 측정값이 있는 배열입니다.

`curve_fit`은 값과 공분산이라는 두 변수를 반환합니다. 변수 values_funcfit은 맞춰진 함수에 대한 매개변수의 최적 값이 있는 배열입니다.

공분산은 맞춰진 매개변수의 공분산을 포함하는 2D 배열입니다. 대각선의 값은 맞춰진 매개변수의 변동을 나타냅니다. 맞춰진 매개변수가 있는 함수의 표준 오차는 다음과 같이 계산할 수 있습니다: `np.sqrt(np.diag(covariance))`. (참고! 이것은 파이썬 속성 과정일 뿐입니다. 데이터 분석, 측정 불확도 및 파이썬에 대한 자세한 정보는 웹에서 쉽게 찾을 수 있습니다).

```python
from scipy.optimize import curve_fit

def fit_function(x,a,b):
        return x**a+b

values, covariance = curve_fit(fit_function, x, y)
```

```python
x_test = np.linspace(0.8*min(x),1.2*max(x),1000) #x-값에 대한 테스트 배열 생성
y_fit = fit_function(x_test,values[0],values[1]) #테스트 배열에 대한 y-값 계산

plt.figure(num='figscatterplot')

plt.plot(x,y,'k.', label='측정값 ')
plt.plot(x_test,y_fit,'r--', label='맞춤')

#시각 자료 향상
plt.xlabel('입력 전압 [V]')
plt.ylabel('출력 전압 [V]')
plt.legend()
plt.grid()
plt.xlim(0,6)

plt.show()

print("a = ",values[0])
print("b = ",values[1])
```
