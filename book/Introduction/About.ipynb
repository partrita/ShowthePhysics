일부는 더 고급 코드를 포함하고 있으며, 가능한 한 간단하게 유지하려고 노력합니다. 아래에는 실행 가능한 코드가 있습니다. 출력은 플롯입니다. 셀을 실행하고, 출력을 보고, 내부의 모든 것을 변경하여 어떻게 작동하는지 볼 수 있습니다...

````python
# 라이브러리 가져오기
import numpy as np
import matplotlib.pyplot as plt

# 일부 데이터
x = np.array([1, 2, 3, 4, 5])   # x에 대한 일부 값
y = 3.14*x**2                   # y에 대한 일부 값

# 그래프 생성
plt.figure()
plt.plot(x, y, 'k.')            # 데이터 포인트를 검은색(k) 점(.)으로 플롯
plt.plot(x,y, 'r--')            # 데이터 포인트를 빨간색(r) 점선(--)으로 플롯

plt.xlabel('$x$[m]')            # 레이블 추가
plt.ylabel('$y$[s]')

plt.grid()                      # 그리드 추가
plt.savefig('plot.png')         # 플롯 저장
plt.show()                      # 플롯 표시
````
